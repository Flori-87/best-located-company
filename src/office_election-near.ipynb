```python
import pandas as pd
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://localhost/datamad0320")
db = client.get_database()
```

```python
cur = db.companiesSF.find({"office":{
    "$geoWithin": {"$geometry": {
             "type" : "Polygon" ,"coordinates":[ [ [ -122.42168 , 37.7683 ], [ -122.42216 , 37.7646 ], [ -122.40962, -37.7669 ] , [ -122.41379 , 37.7714 ] , [ -122.42168 , 37.7683 ]]]}}}})
```

```python
df = pd.DataFrame(cur)
display(df.head(),df.shape)
```

```python
df = df.drop(columns="_id")
df.head()
```

```python
df.to_json("../output/companies_near.json",orient="records")
```

```python
# mas cerca del vegano [-122.42216, 37.7646]
cur = db.companies_near.find({"office": {
         "$near": {
           "$geometry": {"type" : "Point" ,"coordinates":[-122.42216, 37.7646]},
           "$maxDistance": 800}}})
```
