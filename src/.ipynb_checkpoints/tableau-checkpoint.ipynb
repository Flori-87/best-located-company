```python
# Lo primero: query para seleccionar las empresas que contengan office y contengan al menos un elemento
all_companies = db.companies.find({},{"name":1,"offices":1,"total_money_raised":1,"number_of_employees":1,
                                   "founded_year":1,"category_code":1})

#{"offices.latitude":{"$ne":"null"}},{"offices.longitude':  { $ne: null } }]}
#es posible que elimine documentos que no quiero
#{"ipo":{"$exists":True}},{"acquisition":{"$exists":True}},{"total_money_raised":{"$exists":True}}
#{"offices":{"$exists":True}},{"offices":{"$not":{"$size":0}}},
#{"total_money_raised":{"$ne":None}},{"number_of_employees":{"$exists":True}},{"number_of_employees":{"$ne":None}},
                                           #{"ipo":{"$ne":None}},{"acquisition":None}]},
```

```python
df = pd.DataFrame(all_companies)
display(df.head(),df.shape)
```

```python
# convierto los id en formato json
for i in range(len(df["_id"])):
    df["_id"].iloc[i] = dumps(df["_id"].iloc[i])
```

```python
len(df["_id"])
```

```python
# separar los objetos de las listas offices
df = df.explode("offices")
display(df.head(),df.shape)
```

```python
# transform office object into GeoPoint for office
def officeToGeoPoint(row):
    office = row.offices
    if type(office) == dict:
        if 'latitude' in office and 'longitude' in office:
            if office["latitude"] and office["longitude"]:
                return ({
                    "type":"Point",
                    "coordinates":[float(office["longitude"]),float(office["latitude"])]
                },"success")
            else:
                return(None,"Invalid lat and long")
        else:
            return (None,"No lat and long keys in office dict")
    return (None,"No office")
```

```python
cleaned_offices = df.apply(officeToGeoPoint, axis=1, result_type="expand")
cleaned_offices.columns = ["office","state"]
display(cleaned_offices.head(),cleaned_offices.shape)
```

```python
cleaned_offices.state.value_counts()
```

```python
company_processed = pd.concat([df,cleaned_offices], axis=1)
display(company_processed.head(),company_processed.shape)
```

```python
companies_offices = company_processed[company_processed["state"] == "success"]
display(companies_offices.head(),companies_offices.shape)
```

```python
companies_offices2 = companies_offices.drop(columns=["offices","_id"])
display(companies_offices2.head(),companies_offices2.shape)
```

```python
# me lo llevo a tableau
def easyLatLng(row):
    of = row.office
    return {
        "latitude":of["coordinates"][1],
        "longitude":of["coordinates"][0]
    }

# https://api.mongodb.com/python/current/api/bson/objectid.html
companies_offices3 = companies_offices2.apply(easyLatLng, axis=1, result_type="expand")
```

```python
display(companies_offices3.head(),companies_offices3.shape)
```

```python
companies_offices2 = pd.concat([companies_offices2, companies_offices3], axis=1)
display(companies_offices2.head(),companies_offices2.shape)
```

```python
companies_tableau = companies_offices2.drop(columns=["office","state"])
display(companies_tableau.head(),companies_tableau.shape)
```

```python
# Export for tableau
companies_tableau.to_json("../output/companies_tableau.json",orient="records")
```
